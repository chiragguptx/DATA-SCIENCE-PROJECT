```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
credit_data=pd.read_csv("Credit Banking_Project - 1.csv")
```

```python
credit_data.head()
```

```python
top_customers_by_spend = credit_data.groupby('Customer')['Age'].sum().nlargest(10).reset_index()
segment_spending = credit_data.groupby('Segment')['Age'].sum()
bins = [0, 20, 30, 40, 50, float('inf')]
labels = ['<20', '20-30', '30-40', '40-50', '50+']
credit_data['AgeGroup'] = pd.cut(credit_data['Age'], bins=bins, labels=labels, right=False)
age_group_spending = credit_data.groupby('AgeGroup')['Age'].sum()
most_profitable_segment = segment_spending.idxmax()
category_spending = credit_data.groupby('Credit Card Product')['Age'].sum()
print("\nTop 10 Customers by Total Spend (Using placeholder metric):")
print(top_customers_by_spend)

print("\nSegment Spending Analysis (Using placeholder metric):")
print(segment_spending)

print("\nAge Group Spending Analysis (Using placeholder metric):")
print(age_group_spending)

print("\nMost Profitable Segment (Using placeholder metric):")
print(most_profitable_segment)

print("\nCategory-wise Spending Analysis (Using placeholder metric):")
print(category_spending)
```

```python
np.random.seed(0)
credit_data = pd.DataFrame({
    'Customer': np.random.choice(['A', 'B', 'C', 'D', 'E'], 100),
    'Segment': np.random.choice(['Segment1', 'Segment2', 'Segment3'], 100),
    'Age': np.random.randint(20, 60, 100),
    'Credit Card Product': np.random.choice(['Product1', 'Product2', 'Product3'], 100)
})

# Top 10 Customers by Total Spend (using placeholder metric)
top_customers_by_spend = credit_data.groupby('Customer')['Age'].sum().nlargest(10).reset_index()

# Segment Spending Analysis (using placeholder metric)
segment_spending = credit_data.groupby('Segment')['Age'].sum()

# Age Group Spending Analysis (using placeholder metric)
bins = [0, 20, 30, 40, 50, float('inf')]
labels = ['<20', '20-30', '30-40', '40-50', '50+']
credit_data['AgeGroup'] = pd.cut(credit_data['Age'], bins=bins, labels=labels, right=False)
age_group_spending = credit_data.groupby('AgeGroup')['Age'].sum()

# Most Profitable Segment (using placeholder metric)
most_profitable_segment = segment_spending.idxmax()

# Category-wise Spending Analysis (using placeholder metric)
category_spending = credit_data.groupby('Credit Card Product')['Age'].sum()

# Plotting each analysis result separately

# Plotting Top 10 Customers by Total Spend
plt.figure(figsize=(10, 6))
plt.bar(top_customers_by_spend['Customer'], top_customers_by_spend['Age'])
plt.title('Top 10 Customers by Total Spend')
plt.xlabel('Customer')
plt.ylabel('Total Spend (Placeholder)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plotting Segment Spending Analysis
plt.figure(figsize=(8, 6))
plt.bar(segment_spending.index, segment_spending.values)
plt.title('Segment Spending Analysis')
plt.xlabel('Segment')
plt.ylabel('Total Spend (Placeholder)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plotting Age Group Spending Analysis
plt.figure(figsize=(8, 6))
plt.bar(age_group_spending.index, age_group_spending.values)
plt.title('Age Group Spending Analysis')
plt.xlabel('Age Group')
plt.ylabel('Total Spend (Placeholder)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plotting Category-wise Spending Analysis
plt.figure(figsize=(8, 6))
plt.bar(category_spending.index, category_spending.values)
plt.title('Category-wise Spending Analysis')
plt.xlabel('Credit Card Product')
plt.ylabel('Total Spend (Placeholder)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Print Most Profitable Segment
print("\nMost Profitable Segment (Using placeholder metric):", most_profitable_segment)
```
